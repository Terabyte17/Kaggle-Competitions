```python
import os
import gc
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization,Flatten,Conv2D,MaxPooling2D,Input
from keras.utils import to_categorical
```

```python
dirname='/kaggle/input/bengaliai-cv19'
train_data=pd.read_csv(os.path.join(dirname,'train.csv'))
train_data.drop(columns='image_id',inplace=True)
```

```python
train_image_data=pd.read_parquet(os.path.join(dirname,'train_image_data_0.parquet'),engine='pyarrow')
train_image_data.drop(columns='image_id',inplace=True)
```

```python
train_image=np.array(train_image_data).reshape(-1,137,236)
del train_image_data
gc.collect()
```

```python
x=train_image.shape[0]
Y_grapheme=train_data.iloc[:x,0]
Y_vowel=train_data.iloc[:x,1]
Y_consonant=train_data.iloc[:x,2]
```

```python
del train_data
gc.collect()
```

```python
train_images=[]
for i in range(x):
    image=train_image[i]
    image=cv2.resize(image,(64,64),interpolation=cv2.INTER_AREA)
    train_images.append(image)
del train_image
gc.collect()
```

```python
inputs=Input(shape=(64,64,1))

model=Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(64,64,1))(inputs)
model=BatchNormalization()(model)
model=Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Conv2D(filters=32,kernel_size=(5,5),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=MaxPooling2D(pool_size=(2, 2),strides=2)(model)

model=Conv2D(filters=64,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Conv2D(filters=64,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Conv2D(filters=64,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=MaxPooling2D(pool_size=(2, 2),strides=2)(model)
model=Dropout(0.3)(model)

model=Conv2D(filters=128,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Conv2D(filters=128,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Conv2D(filters=128,kernel_size=(3,3),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=MaxPooling2D(pool_size=(2, 2),strides=2)(model)
model=Dropout(0.3)(model)

model=Conv2D(filters=256,kernel_size=(5,5),padding='same',activation='relu')(model)
model=BatchNormalization()(model)
model=Dropout(0.3)(model)

model=Flatten()(model)
model=Dense(1024,activation = "relu")(model)
model=Dropout(0.3)(model)
dense=Dense(512, activation = "relu")(model)

grapheme=Dense(168,activation='softmax',name='grapheme')(dense)
vowel=Dense(11,activation='softmax',name='vowel')(dense)
consonant=Dense(7,activation='softmax',name='consonant')(dense)


model=Model(inputs=inputs,outputs=[grapheme,vowel,consonant])
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
Y_grapheme=to_categorical(Y_grapheme,num_classes=168)
Y_vowel=to_categorical(Y_vowel,num_classes=11)
Y_consonant=to_categorical(Y_consonant,num_classes=7)
```

```python
train_images=np.array(train_images)
```

```python
train_images=train_images.reshape(-1,64,64,1)
```

```python
model.fit(train_images,[Y_grapheme,Y_vowel,Y_consonant],batch_size=32,epochs=20,verbose=1,validation_split=0.1)
```

```python
del Y_grapheme
del Y_vowel
del Y_consonant
del train_images
gc.collect()
```
