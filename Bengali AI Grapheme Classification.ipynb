```python
import os
import gc
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential,Model,load_model
from tensorflow.keras.layers import Dense,Input,Flatten,BatchNormalization,Dropout,Conv2D,MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

```python
dirname_1='/kaggle/input/bengaliai-cv19'
dirname_2='/kaggle/input/model10'
```

```python
submission=pd.read_csv(os.path.join(dirname_1,'sample_submission.csv'))
```

```python
test_image=pd.read_parquet(os.path.join(dirname_1,'test_image_data_'+str(0)+'.parquet'),engine='pyarrow')
test_image.drop(columns=['image_id'],inplace=True)
```

```python
test_image=np.array(test_image).reshape(-1,137,236,1)
```

```python
test_image_data=[]
for i in range(test_image.shape[0]):
    image=test_image[i]
    image=cv2.resize(image,(64,64),interpolation=cv2.INTER_AREA)
    test_image_data.append(image)
del test_image
gc.collect()
```

```python
test_image_data=np.array(test_image_data)
```

```python
test_image_data=test_image_data.reshape(-1,64,64,1)
```

```python
model=Sequential()
model=load_model(os.path.join(dirname_2,'new_model10_weights.h5'))
```

```python
count=0
y=model.predict(test_image_data)
graphemes=np.argmax(y[0],axis=1)
vowels=np.argmax(y[1],axis=1)
consonants=np.argmax(y[2],axis=1)
for i in range(len(test_image_data)):
    submission.loc[count,'target']=consonants[i]
    count+=1
    submission.loc[count,'target']=graphemes[i]
    count+=1
    submission.loc[count,'target']=vowels[i]
    count+=1
del y
del graphemes
del vowels
del consonants
del test_image_data
```
